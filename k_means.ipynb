K-Means:无监督聚类算法：
- 选K:人工指定要分成几类
- 随机选质心
- 归类
- 更新质心
- 重复3-4次


## K-Means优化目标（簇内平方和误差 SSE）
$$J = \sum_{k=1}^{K} \sum_{x_i \in C_k} \| x_i - \mu_k \|_2^2$$



#### 算法步骤：
- 随机选K个点作为初始聚类中心
- 计算所有点到K个中心的距离，把每个点分给最近的中心
- 更新：用每个簇里所有点的平均值，计算新的聚类中心
- 重复2、3步，直到中心不再变化（收敛）


```python
import numpy as np
from typing import List, Tuple

class Solution:
    def kMeans(
        self,
        n: int,
        k: int,
        max_iter: int,
        nums: List[List[float]]
    ) -> Tuple[List[List[str]], List[int]]:
        # 转为 numpy 数组
        points = np.array(nums, dtype=float)          # shape: (n, 2)

        # 初始化聚类中心：直接取前 k 个点
        centers = points[:k].copy()                  # shape: (k, 2)

        # 初始化每个点所属的簇标签
        labels = np.full(n, -1, dtype=int)

        for _ in range(max_iter):
            # 第一步：计算每个点到每个中心的平方距离
            # dists[i, j] 表示第 i 个点到第 j 个中心的平方距离
            dists = np.sum((points[:, None, :] - centers[None, :, :]) ** 2, axis=2)

            # 为每个点分配最近的簇
            new_labels = np.argmin(dists, axis=1)

            # 如果本轮没有任何标签变化，说明已经收敛
            if np.array_equal(labels, new_labels):
                break

            labels = new_labels

            # 第二步：根据新标签更新每个簇的中心
            for j in range(k):
                cluster_points = points[labels == j]
                if len(cluster_points) > 0:
                    centers[j] = np.mean(cluster_points, axis=0)
                # 如果某个簇没有点，保持原中心不变

        # 把中心转换成固定 4 位小数的字符串
        centers_str = [[f"{x:.4f}", f"{y:.4f}"] for x, y in centers]

        return centers_str, labels.tolist()
```
